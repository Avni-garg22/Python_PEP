```python
'''most of the Matplotlib utilities lies under the pyplot
submodule, and are usually imported under the plt alias: '''

import matplotlib.pyplot as plt
import numpy as np
```

```python
#Draw a line in a digram from position(0,0) to position(6,250):

xpoints=np.array([0,6])
ypoints=np.array([0,250])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
''''PLotting x and y points:
parameter 1 is an array containing the points on the x-axis.
parameter 2 is an array containing the points on the y-axis.
'''

#draw a line in a diagram from positions (1,3) to position (8,10):

xpoint=np.array([1,8])
ypoints=np.array([3,10])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
#plotting without line

xpoint=np.array([1,8])
ypoints=np.array([3,10])

plt.plot(xpoints, ypoints,'o')
plt.show()
```

```python
#draw a line in a diagram from position (1,3) to (2,8) then to (6,1) and finally to poisition (8,10)

xpoints=np.array([1,2,6,8])
ypoints=np.array([3,8,1,10])

plt.plot(xpoints,ypoints)
plt.show()
```

```python
#plotting without x-axis points:

ypoints=np.array([3,8,1,10,5,7])
plt.plot(ypoints)
plt.show()
```

```python
#Mark each point with a circle(or anything)
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o')

plt.show()

ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='v')

plt.show()
```

```python
ypoints=np.array([3,6,9,10])

plt.plot(ypoints,'o:r')
plt.show()


ypoints=np.array([3,6,9,1,10])

plt.plot(ypoints,'o:b')
plt.show()
```

```python
ypoints=np.array([3,6,9,10])

plt.plot(ypoints,'o:r', ms=20, mec='b')
plt.show()

ypoints=np.array([3,6,9,1,2,10])

plt.plot(ypoints,'o:g',15)
plt.show()
```

```python
#Set font properties for the title and labels:

x=np.array([80,85,90,95,100,105,110])
y=np.array([80,85,90,95,100,105,110])

font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x,y)
plt.show()
```

```python
#draw 4 bars:

x= np.array(["A","B","C","D"])
y= np.array([3,8,1,10])

plt.bar(x,y)
plt.show()
```

```python
''' A histogram is a graph showing frequency distributions. 
It is a graph showing a number of observations within each given interval
'''

x=np.random.normal(170,10,250)

plt.hist(x)
plt.show()
```

```python
#Start the first wedge at 90 degrees:

y =np.array([35,25,25,17])
mylabels=["Apples","Bananas","Cherries","Dates"]

plt.pie(y, labels=mylabels, startangle=90)
plt.show()
```
